```python
def dms_to_dd(graus, minutos, segundos, hemisferio):
    """
    Converte coordenadas em Graus, Minutos e Segundos (DMS)
    para Graus Decimais (DD).

    hemisferio:
        - 'N' ou 'S' para latitude
        - 'E' ou 'W' para longitude
    """

    decimal = graus + minutos/60 + segundos/3600

    # Hemisférios negativos na Cartografia:
    if hemisferio.upper() in ["S", "W"]:
        decimal *= -1

    return decimal


print("=== Conversor de Coordenadas DMS → DD ===")

# Entrada do usuário
g = float(input("Graus: "))
m = float(input("Minutos: "))
s = float(input("Segundos: "))
h = input("Hemisfério (N/S/E/W): ")

resultado = dms_to_dd(g, m, s, h)

print(f"\nCoordenada em graus decimais: {resultado}")
```
